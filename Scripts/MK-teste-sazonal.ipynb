Aplicação e plot do teste Mann-Kendall modificado (Yue and Wang) sobre dados de anomalia padronizada sazonal nas 3 regiões de estudo

```python
import numpy as np
import pandas as pd
import pymannkendall as mk
import matplotlib.pyplot as plt
import statsmodels.api as sm
%matplotlib inline
```

```python
arquivo1 = '/home/beatriz/Desktop/UFRJ/períodos/2022.2/f-clima/trabalho/sazonal/areas/A1_sazonal.csv'
df1 = pd.read_csv(arquivo1,na_values='NaN',sep=',')
```

Teste de significancia

```python
# A1
data = df1['f latente']
res = mk.yue_wang_modification_test(data)
print(res.h)

data2 = df1['f sensível']
res = mk.yue_wang_modification_test(data2)
print(res.h)

data3 = df1['ciclogenese']
res = mk.yue_wang_modification_test(data3)
print(res.h)
```

```python
# A1
data = df1['f latente']

fig, ax = plt.subplots(3,1,figsize=(14, 12))

res = mk.yue_wang_modification_test(data)
trend_line = np.arange(len(data)) * res.slope + res.intercept

ax[0].plot(data,color='b')
ax[0].plot(data.index, trend_line,color='black', linewidth=3)
ax[0].legend(['FCL', 'Tendência'],fontsize=13)
ax[0].grid(True)
#ax[0].set_ylim([-0.5, 0.5])
ax[0].set_xticklabels(df1['ano'][0::15], rotation=0,fontsize=14)
ax[0].set_title('A1',fontsize=15,loc='center')
ax[0].set_title('a)',fontsize=12,loc='left')

data2 = df1['f sensível']
res = mk.yue_wang_modification_test(data2)
trend_line = np.arange(len(data2)) * res.slope + res.intercept

ax[1].plot(data2,color='g')
ax[1].plot(data2.index, trend_line,color='black', linewidth=3)
ax[1].legend(['FCS', 'Tendência'],fontsize=13)
ax[1].grid(True)
#ax[1].set_ylim([-1.0, 1.0])
ax[1].set_xticklabels(df1['ano'][0::15], rotation=0,fontsize=14)
ax[1].set_title('b)',fontsize=12,loc='left')

data3 = df1['ciclogenese']
res = mk.yue_wang_modification_test(data3)
trend_line = np.arange(len(data3)) * res.slope + res.intercept

ax[2].plot(data3,color='purple')
#ax[2].plot(data3.index, trend_line,color='black', linewidth=3)
ax[2].legend(['Ciclogenese'],fontsize=13)
ax[2].grid(True)
#ax[2].set_ylim([-1.5, 2.5])
ax[2].set_xticklabels(df1['ano'][0::15], rotation=0,fontsize=14)
ax[2].set_title('c)',fontsize=12,loc='left')

plt.savefig('/home/beatriz/Desktop/UFRJ/períodos/2022.2/f-clima/trabalho/sazonal/A1_trend_.jpg', bbox_inches='tight')
```
